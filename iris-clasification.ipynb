# Loading the [iris dataset](https://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html) 

```python
from sklearn import datasets

iris = datasets.load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target

df_iris.head()
```

# Creating a data and training set

```python
from sklearn.model_selection import train_test_split

X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

Exporting as txt

```python
df_train = pd.DataFrame(np.concatenate((X_train, y_train.reshape(-1, 1)), axis=1))
df_test = pd.DataFrame(np.concatenate((X_test, y_test.reshape(-1, 1)), axis=1))

df_train.to_csv('train.txt', sep=" ", index=False, header=False)
df_test.to_csv('test.txt', sep=" ", index=False, header=False)
```
